```python
import pandas as pd

df3= pd.read_csv(r"D:\ecommerceDataset.csv",names= ["category","description"],header=None)
print(df3.columns)
```

```python
df3.dropna(inplace=True)
```

```python
df3.shape
```

```python
df3.category.replace("Clothing & Accessories","Clothing_Accessories",inplace=True)
df3.category.value_counts()
```

```python
df3['category']= "__label__"+df3['category'].astype(str)
df3.head()
```

```python
df3['category_description']=  df3['category']+ " " +df3['description']
df3
```

```python
df3['category_description'][0]

#in this the description contains the special characters hence we need to remove it for ease of classification for the fasttext model
```

```python
import re

def preprocess(text):
    text=re.sub(r'[^\w\s\']',' ',text)
    text=re.sub(r' +',' ' ,text)
    return text.strip().lower()
#need to convert to = viki's bookcase bookshelf 3 shelf shelve white hi

```

```python

preprocess("VIKI's | Bookcase/Bookshelf (3-Shelf/Shelve, White) | ? . hi")
```

```python
df3['category_description']=df3['category_description'].apply(preprocess)
```

```python
df3['category_description'].head(5)
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df3,test_size=0.2)
```

```python
train.shape
```

```python
test.shape
```

```python
train.to_csv("ecommerce.train",columns = ["category_description"],index=False, header= False)
test.to_csv("ecommerce.test",columns = ["category_description"],index=False,header=False)
```

```python
import fasttext
model = fasttext.train_supervised(input = "ecommerce.train")
model.test("ecommerce.test")
# (size of sample , precision , recall)
```
